Now I want to select the number of bins to plot the distribution of the run durations using several methods and testing which method fits the distribution best.

```python
# # Square Root Rule
# num_bins = int(np.sqrt(len(runtime_df['run_duration'])))

#Freedman-Diaconis Rule
# iqr = runtime_df['run_duration'].quantile(0.75) - runtime_df['run_duration'].quantile(0.25)
# bin_width = 2 * iqr * (len(runtime_df['run_duration']) ** (-1 / 3))
# num_bins = int((runtime_df['run_duration'].max() - runtime_df['run_duration'].min()) / bin_width)

# Sturges' Rule
# num_bins = int(np.ceil(np.log2(len(runtime_df['run_duration'])) + 1))

# Scott's Rule
bin_width = 3.5 * runtime_df['run_duration'].std() / (len(runtime_df['run_duration']) ** (1 / 3))
num_bins = int((runtime_df['run_duration'].max() - runtime_df['run_duration'].min()) / bin_width)
print(num_bins)
```

```python
# Histogram of run durations
plt.hist(runtime_df['run_duration'].dt.total_seconds() / 3600, bins=num_bins)
plt.xlabel('Run Duration (hours)')
plt.ylabel('Frequency')
plt.title('Distribution of Production Run Durations')
plt.show()
```

We then merge sensor data with the production runs based on timestamp. Then we filter the sensor data mearsured between the start time and end time of each production run. Then aggregate the rows where the sensor timestamp falls within the start and end times of production runs using mean() and sum (). I used mean aggregation for the temperature, pressure data to understand the average behavior of each sensor's readings over a production run and changes that may occur and the sum aggregation to capture the cummulative power consumption during the production run.

```python
# converting sensor data timestamp to datetime object
sensor_df["timestamp"] = pd.to_datetime(sensor_df["timestamp"])

# Merge sensor data with production runs based on timestamps
merged_data = pd.merge_asof(sensor_df.sort_values('timestamp'),
                            runtime_df[['start_time', 'end_time']].sort_values('start_time'),
                            left_on='timestamp',
                            right_on='start_time',
                            direction='backward')

# Filter rows where sensor timestamp falls within production run start and end times
filtered_data = merged_data[
    (merged_data['timestamp'] >= merged_data['start_time']) &
    (merged_data['timestamp'] <= merged_data['end_time'])
]

# Aggregate sensor data based on mean and sum
aggregated_data = filtered_data.groupby(['start_time', 'end_time']).agg({
    'building_temperature': 'mean',
    'input_power1': 'sum',
    'input_power2': 'sum',
    'input_power3': 'sum',
    'input_power4': 'sum',
    'input_power5': 'sum',
    'output_power1': 'sum',
    'output_power2': 'sum',
    'output_power3': 'sum',
    'output_power4': 'sum',
    'output_power5': 'sum',
    'pressure1': 'mean',
    'pressure2': 'mean',
    'pressure3': 'mean',
    'pressure4': 'mean',
    'pressure5': 'mean',
    'temperature1': 'mean',
    'temperature2': 'mean',
    'temperature3': 'mean',
    'temperature4': 'mean',
    'temperature5': 'mean'
}).reset_index()
```

```python
# Visualizing the aggregated data
aggregated_data
```

```python
# Extract necessary columns from runtime_df
extra_column = runtime_df[['start_time', 'failure_time']]

# Merge additional_column with aggregated_data based on start_time
aggregated_data = pd.merge(aggregated_data, extra_column, on='start_time', how='left')

aggregated_data.head(10)
```

Now I am tring to prepare the sensor data to extract important features relevant in predicting the failure events, by making it look like to the original sensor dataset. We start by droping the start time and the end time. Next we add an extra column named 'failure time' to be used to create a target class named 'system failed' which will have boolean values that will tell if a system failed during a production run or not

```python
# I drop the start time and end time column obtained from previous aggregation
aggregated_data.drop(['start_time', 'end_time'], axis=1, inplace=True)
aggregated_data.head(10)
```

```python
# Check for NaN values using pd.isnull() to create the 'system_failed' column
aggregated_data['system_failed'] = ~pd.isnull(aggregated_data['failure_time'])
```

Finally drop the failure time column after creating the target class system failed

```python
aggregated_data.drop(['failure_time'], axis=1, inplace=True)
aggregated_data.head(10)
```

Now we want to extract the relevant data that are most relevant to predict the the failure events. by applying a feaw methods such as:


1.   Corelation analysis
2.   One-way ANOVA





1.   Now lets look at the correlation analysis between the independent features and then between indepentent variables and the target variable


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculating the correlation matrix
corr_matrix = aggregated_data.corr()

# Applying precision formatting to the correlation matrix
corr_matrix = corr_matrix.round(2)  # to 2 decimal places

# Create a heatmap plot using Seaborn with formatted values
plt.figure(figsize=(12, 8))  # Seting the size of the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Rotate x-axis labels (features) by 45 degrees and adjust alignment
plt.xticks(rotation=45, ha='right')

plt.title('Heatmap of Feature Correlation (Precision: 2)')
plt.tight_layout()
plt.show()
```


2.   Using One-way ANOVA to check for the significant features



```python
from scipy.stats import f_oneway

# Loop through each feature and perform ANOVA
feature_scores = {}
X = aggregated_data.iloc[:, :-1]
y = aggregated_data['system_failed']
for col in X.columns:
    groups = [X[col][y == level] for level in y.unique()]
    f_stat, p_value = f_oneway(*groups)
    feature_scores[col] = p_value

# Set significance level (alpha)
alpha = 0.05

# Select features with p-value less than alpha (significant features)
significant_features = [feat for feat, pval in feature_scores.items() if pval < alpha]
# list of the most significant features
significant_features
```
